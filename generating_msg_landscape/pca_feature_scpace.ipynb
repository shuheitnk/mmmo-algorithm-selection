```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from matplotlib import rcParams
from pathlib import Path

rcParams["text.usetex"] = False
rcParams["font.family"] = "arial"
rcParams["mathtext.fontset"] = "cm"
rcParams["pdf.fonttype"] = 42
rcParams["ps.fonttype"] = 42
rcParams["figure.dpi"] = 300
rcParams["savefig.dpi"] = 300

d = 10
data_path = Path(f"../generated_msg/combined_{d}d_median_stable.csv")
df = pd.read_csv(data_path)

feature_cols = [c for c in df.columns if c not in ["function_id", "instance_id", "seed"]]
X = df[feature_cols].values
y = df["function_id"].values
unique_ids = sorted(df["function_id"].unique())


bbob_ids = list(range(1, 25))
bbob_color = "#aaaaaa"

category_colors = {
    121: "#1f77b4",
    112: "#ff7f0e",
    111: "#2ca02c",
    122: "#d62728",
    211: "#9467bd",
    212: "#8c564b",
    221: "#e377c2",
    222: "#17becf",
}

category_labels = {
    111: "min_min_min",
    112: "min_min_max",
    121: "min_max_min",
    122: "min_max_max",
    211: "max_min_min",
    212: "max_min_max",
    221: "max_max_min",
    222: "max_max_max",
}

X_std = StandardScaler().fit_transform(X)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_std)
var_ratio = pca.explained_variance_ratio_ * 100

fig, ax = plt.subplots(figsize=(7, 6.5))

mask_bbob = df["function_id"].isin(bbob_ids)
ax.scatter(
    X_pca[mask_bbob, 0],
    X_pca[mask_bbob, 1],
    c=bbob_color,
    s=70,
    alpha=0.9,
    edgecolor="none",
    label="BBOB"
)

for fid in category_colors:
    mask = df["function_id"] == fid
    ax.scatter(
        X_pca[mask, 0],
        X_pca[mask, 1],
        c=category_colors[fid],
        s=70,
        alpha=0.8,
        edgecolor="k",
        linewidth=1,
        label=category_labels[fid]
    )

for fid in category_colors:
    df_sub = df[df["function_id"] == fid].sort_values("instance_id")
    idx = df_sub.index
    ax.plot(
        X_pca[idx, 0],
        X_pca[idx, 1],
        color=category_colors[fid],
        linewidth=2,
        alpha=1.0,
        zorder=1
    )

ax.set_xlabel(f"Component 1 ({var_ratio[0]:.1f}%)", fontsize=25)
ax.set_ylabel(f"Component 2 ({var_ratio[1]:.1f}%)", fontsize=25)
ax.set_title(rf"{d}d Benchmark Problems", fontsize=30, pad=15)
ax.grid(alpha=0.3)

for spine in ax.spines.values():
    spine.set_linewidth(1.5)

output_dir = Path("../generated_msg")
output_dir.mkdir(exist_ok=True)

output_path = output_dir / f"pca_ela_{d}d_msg.svg"
plt.savefig(output_path, format="svg", bbox_inches="tight", transparent=True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from matplotlib import rcParams
from pathlib import Path

rcParams["text.usetex"] = False
rcParams["font.family"] = "arial"
rcParams["mathtext.fontset"] = "cm"
rcParams["pdf.fonttype"] = 42
rcParams["ps.fonttype"] = 42
rcParams["figure.dpi"] = 300
rcParams["savefig.dpi"] = 300



d = 10
data_path = Path(f"../generated_msg/combined_{d}d_median_stable.csv")
df = pd.read_csv(data_path)

feature_cols = [c for c in df.columns if c not in ["function_id", "instance_id", "seed"]]
X = df[feature_cols].values
y = df["function_id"].values
unique_ids = sorted(df["function_id"].unique())


bbob_ids = list(range(1, 25))

bbob_colors = [
    "#cce5ff", "#99ccff", "#66b2ff", "#3399ff", "#0066cc",
    "#ccffcc", "#99ff99", "#66ff66", "#00cc00",
    "#ffcce6", "#ff99cc", "#ff66b2", "#ff3399", "#e60073",
    "#fff9cc", "#fff399", "#ffee66", "#ffdd33", "#e6cc00",
    "#ffcccc", "#ff9999", "#ff6666", "#ff3333", "#cc0000",
]

msg_ids = [111, 112, 121, 122, 211, 212, 221, 222]


X_std = StandardScaler().fit_transform(X)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_std)

var_ratio = pca.explained_variance_ratio_ * 100


fig, ax = plt.subplots(figsize=(7, 6.5))

# BBOB functions
for fid in bbob_ids:
    mask = df["function_id"] == fid
    ax.scatter(
        X_pca[mask, 0],
        X_pca[mask, 1],
        c=bbob_colors[fid - 1],
        s=70,
        alpha=0.8,
        edgecolor="k",
        linewidth=1,
        zorder=2,
    )

for fid in msg_ids:
    mask = df["function_id"] == fid
    ax.scatter(
        X_pca[mask, 0],
        X_pca[mask, 1],
        c="#aaaaaa",
        s=70,
        alpha=0.9,
        edgecolor="none",
        zorder=1,
    )

ax.set_xlabel(f"Component 1 ({var_ratio[0]:.1f}%)", fontsize=25)
ax.set_ylabel(f"Component 2 ({var_ratio[1]:.1f}%)", fontsize=25)
ax.set_title(rf"{d}d Benchmark Problems", fontsize=30, pad=15)
ax.grid(alpha=0.3)

for spine in ax.spines.values():
    spine.set_linewidth(1.5)


output_path = Path("../generated_msg") / f"pca_ela_{d}d_bbob.svg"
plt.savefig(output_path, format="svg", bbox_inches="tight", transparent=True)
plt.show()
```
